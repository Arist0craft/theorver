# A/B тестирование. И с чем его едят.
Два Jupyter Notebook'а будут написаны с целью рассказать об A/B тестах. Они часто используются в социологических, психологических и маркетинговых исследованиях, для оценки влияния того или иного изменения во время эксперимента.
Начнем с понятий:

## 1. Основные понятия:
__A/B тестирование или Split тестирование__ - метод, при котором генеральная совокупность делится на две или более выборок. Первая выборка A - контрольная, другие выборки B - тестовые, на которых проверяют гипотезы.  
  
__Гипотеза__ - предположение о влиянии некоторых факторов на целивые метрики.  
  
Например, у нас есть сайт, на котором человек может купить товар. На одном из этапов покупки, всплывает окно с длинным текстом, который явно мешает покупке. Обозначим за целевую метрику конверсии из новых пользователей сайта в покупателей или просто количество покупок. Выдвенем гипотезу, что, убрав popup (всплывающее окно), мы поднимем количество конверсий на сайте. Далее, условно, делаем версию страницы, где не будет всплывать раздражающий popup, делим аудиторию с помощью скрпита, показывающего в 50% случаев страницу без popup, некоторое время ждем, пока соберется статистика, после чего оцениваем результат. Поднялись ли конверсии, упали или остались на прежнем уровне? Стала ли причиной изменения конверсий наша гипотеза?
