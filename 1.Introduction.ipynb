## 1. Основные понятия.

Математическая статистика плотно связана с теорией вероятностью, приведу ассоциацию с физикой, где есть физики практики, ставящие эксперименты и получающие результаты в экспериментах, а также физики теоретики, которые строят теорию и модели, позволяющие точно описать процессы, проходящие во время эксперимента.  
Однако в физик как теоретики, так и практики говорят на одном языке, пользуются теми же понятиями, поэтому для изучения статистики и теории вероятности нам тоже потребуются основные понятия из теории вероятности и математической статистики.

__Исход__ - исход или результат эксперимента, например, эксперимент: бросок монеты, исход: решка. Чаще всего обозначается, как $\omega$

$$\Omega = \{ \omega_{1}, ..., \omega_{i} \} - множество\ всех\ исходов$$
Сразу стоит заметить, что в теории вероятности нам необходимо знать все исходы, поэтому множество всех исходов является фундаментальным понятием, однако статистика сильно отличается, ведь в реальности мы вряд ли сможем узнать все исходы.

$$A =  \{ \omega_{l}, ..., \omega_{m} \} \subset \Omega - случайное\ событие$$
__Случайное событие__ - набор исходов из множества всех исходов, считается, что если в эксперменте был получен исход, заранее неизвестно какой из всех, $\omega_{j} \in A$, то произошло случайное событие $A$

__Случайной величиной__ - переменная величина, которая в
результате испытания в зависимости от исхода принимает одно из возможного
множества своих значений, причем заранее неизвестно, какое именно.  
Случайные величины в статистике чаще всего обозначают латинскими заглавными буквами $X, Y, Z$, а их значения, соответственно, маленькими $x_{1}, x_{2}, z_{3}$. В теории вероятностей чаще всего встречается обозначение $\xi$ для отражения более абстрактных случаев случайных величин. Мы будем рассматривать величины, которые относятся к $\mathbb{R}$, то есть к множеству рациональных чисел.    
По определению:
$$ X: \Omega \rightarrow \mathbb{R} $$
$$ \xi: \Omega \rightarrow \mathbb{R} $$
  
Например, обозначим за $\omega_{1} = решка,\ \omega_{2} = решка$, тогда $\xi (\omega_{1}) = 1,\ \xi(\omega_{2}) = 0$

## 2. Вероятность.
Что такое вероятность и с чем её едят?  


### 2.1 Классическое определение вероятности.  
Вероятность исхода:
$$P(\omega_{i}) = \frac{1}{n}, где\ n = \vert \Omega \vert$$  
Вероятность события A:
$$P(A) = \frac{m}{n}, где\ m = \vert A \vert, n = \vert \Omega \vert$$  
$m$ - благоприятные исходы, те, которые сопутстувуют событию A, $n$ - все исходы.  
Вероятность любого события принимает значения $0 <= P(A) <= 1$, где 0 - событие невозможно, а 1 - событие произойдет.  
Также вероятность измеряют в процентах, достаточно домножить её на 100, поэтому можно сказать, что вероятность - процент испытаний, в которых МОЖЕТ, но необязательно, произойти событие A.

### 2.3 Статистическое определение вероятности.
Относительная частота события A:
$$W(A) = \frac{m}{n}, где\ m - количество\ испытаний, в\ которых\ событие\ A\ наступило, n - количество\ всех\ испытаний$$  
За вероятность в статистике принимают относистельную частоту события, соответственно $P(A) = W(A)$
